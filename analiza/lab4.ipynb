```python
import numpy as np
from math import *

arr_x = np.arange(-10, 10, 0.1)
arr_y = [sin(4 * i) for i in arr_x]
import matplotlib.pyplot as plt

plt.plot(arr_x, arr_y, 'm*--')




```

```python
arr_x = np.arange(-10, 10, 1)
arr_y = [-5 * i + 10 for i in arr_x]
import matplotlib.pyplot as plt
plt.axvline()
plt.axhline()
plt.plot(arr_x, arr_y, 'mH:')
arr_x = np.arange(-10, 10, 1)
arr_y = [-5 * sin(i) + 10 for i in arr_x]
plt.subplot(arr_x, arr_y, "yD--")
```
